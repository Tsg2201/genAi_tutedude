```python
df=pd.read_csv('hr_dataset.csv')
df.head()
```

```python
# Group salary by year
yearly_salary = df.groupby('Year')['Salary'].mean()

plt.plot(yearly_salary.index, yearly_salary.values)

plt.title("Average Salary Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Average Salary")

plt.show()
```
