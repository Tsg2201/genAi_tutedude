```python
df=pd.read_csv('hr_dataset.csv')
df.head()
```

```python
dept_salary = df.groupby(['Department', 'Job_Level'])['Salary'].sum().unstack()

bottom = np.zeros(len(dept_salary))

for col in dept_salary.columns:
    plt.bar(dept_salary.index, dept_salary[col], bottom=bottom, label=col)
    bottom += dept_salary[col].values

plt.title("Stacked Salary by Department & Job Level")
plt.xlabel("Department")
plt.ylabel("Total Salary")

plt.legend()
plt.xticks(rotation=45)
plt.show()
```
