```python
df=pd.read_csv('hr_dataset.csv')
df.head()
```

```python
dept_counts = df['Department'].value_counts()

# Vertical Bar
plt.bar(dept_counts.index, dept_counts.values)
plt.title("Employees per Department (Vertical)")
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Horizontal Bar
plt.barh(dept_counts.index, dept_counts.values)
plt.title("Employees per Department (Horizontal)")
plt.xlabel("Count")
plt.ylabel("Department")
plt.show()
```
