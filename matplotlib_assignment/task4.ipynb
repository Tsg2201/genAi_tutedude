```python
df=pd.read_csv('hr_dataset.csv')
df.head()
```

```python
grouped = df.groupby(['Year', 'Job_Level'])['Salary'].mean().unstack()

years = grouped.index
x = np.arange(len(years))
width = 0.25

levels = grouped.columns

for i, level in enumerate(levels):
    plt.bar(x + i*width, grouped[level], width, label=level)

plt.xticks(x + width, years)

plt.title("Average Salary by Year & Job Level")
plt.xlabel("Year")
plt.ylabel("Salary")

plt.legend()
plt.show()
```
