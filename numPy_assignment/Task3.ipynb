```python
values=np.array([2,4,6,8,10]) #initializing an array of numbers using array function
print("Values array:", values)
```

```python
#Square root of each element in the array using np.sqrt() function
sqrt_values=np.sqrt(values) 
print("Square root of each element in Values array:", sqrt_values)

#Exponential of each element in the array using np.exp() function
exp_values=np.exp(values)
print("Exponential of each element in Values array:", exp_values)

#natural logarithm of each element in the array using np.log() function
log_values=np.log(values)
print("Natural logarithm of each element in Values array:", log_values)

#Sum of all elements in the array using np.sum() function
sum_values=np.sum(values)
print("Sum of all elements in Values array:", sum_values)
#Cumulative sum of the elements in the array using np.cumsum() function
cumsum_values=np.cumsum(values)
print("Cumulative sum of elements in Values array:", cumsum_values)
```
