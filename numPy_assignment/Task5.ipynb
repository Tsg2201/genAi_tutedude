```python
marks=np.array([78,85, 90, 66, 72, 88,95,60]) #initializing a 1D array using array function
print("Marks array:\n", marks)
```

```python
#Mean of the marks
mean_marks=np.mean(marks)
print("Mean of the marks:", mean_marks)

#Median of the marks
median_marks=np.median(marks)
print("Median of the marks:", median_marks)

#Variance of the marks
variance_marks=np.var(marks)
print("Variance of the marks:", variance_marks)

#Standard deviation of the marks
std_dev_marks=np.std(marks)
print("Standard Deviation of the marks:", std_dev_marks)

#Minimum and Maximum marks
min_marks=np.min(marks)
max_marks=np.max(marks)
print("Minimum marks:", min_marks)
print("Maximum marks:", max_marks)

#Range of the marks
range_marks=max_marks - min_marks
print("Range(max-min):", range_marks)
```
