```python
sales = np.array([1200, 1500, 900, 2000, 1800,1700,1600 ]) #initializing a 1D array using array function
print("Sales array:\n", sales)
```

```python
#Total weekly sales
total_sales = np.sum(sales)
print("Total weekly sales:", total_sales)

#Average daily sales
average_sales = np.mean(sales)
print("Average daily sales:", average_sales)

#Highest and lowest sales day
highest_sales = np.max(sales)
lowest_sales = np.min(sales)
print("Highest sales day:", highest_sales)
print("Lowest sales day:", lowest_sales)

#Standard deviation of sales
std_dev_sales = np.std(sales)
print("Standard deviation of sales:", std_dev_sales)

#Number of days with sales above average
above_average_days = np.sum(sales > average_sales)
print("Number of days with sales above average:", above_average_days)
```
