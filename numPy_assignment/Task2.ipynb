```python
A=np.array([10,20,30,40]) #creating an array of numbers using array function
B=np.array([1,2,3,4])
```

```python
sum_arr=A+B #adding two arrays A and B
print("Sum of arrays:", sum_arr)

diff_arr=A-B #subtracting two arrays A and B
print("Difference of arrays:", diff_arr)

mul_arr=A*B #multiplying two arrays A and B
print("Multiplication of arrays:", mul_arr) 

div_arr=A/B #dividing two arrays A and B
print("Division of arrays:", div_arr)

pow_arr=A**2 #raising array A to the power of 2
print("Power of arrays:", pow_arr)
```

```python
#USING .add() function to add two arrays A and B
sum_arr_func=np.add(A,B)
print("Sum of arrays using function:", sum_arr_func)

#USING .subtract() function to subtract two arrays A and B
diff_arr_func=np.subtract(A,B)  
print("Difference of arrays using function:", diff_arr_func)
```
