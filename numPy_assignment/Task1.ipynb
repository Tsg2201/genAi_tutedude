```python
array1=np.arange(1,11) #creating an array of numbers from 1 to 10 using arange function
print(array1)
```

```python
array2=np.arange(1,10).reshape(3,3) #creating an array of numbers from 1 to 9 using arange function and shape function to reshape it into a 3x3 array
print(array2)
```

```python
array3=np.array([10,20,30,40,50]) #creating an array of numbers using array function
print(array3)
```

```python
print("Shape of array1:", array1.shape) #to check the shape of the array1
print("Shape of array2:", array2.shape) #to check the shape of the array2
print("Shape of array3:", array3.shape) #to check the shape of the array3
```

```python
print("Datatype of array1:", array1.dtype) #to check the data type of the array1
print("Datatype of array2:", array2.dtype) #to check the data type of the array2
print("Datatype of array3:", array3.dtype) #to check the data type of the array3
```
