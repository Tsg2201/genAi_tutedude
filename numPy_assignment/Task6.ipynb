```python
marks=np.array([78,85, 90, 66, 72, 88,95,60]) #initializing a 1D array using array function same as Task5
print("Marks array:\n", marks)
```

```python
#Sorting the marks array in ascending order
sorted_marks = np.sort(marks)
print("Sorted Marks array:\n", sorted_marks)

#25th Percentile of marks array
percentile_25 = np.percentile(marks, 25)
print("25th Percentile of marks array:", percentile_25)

#50th Percentile of marks array
percentile_50 = np.percentile(marks, 50)    
print("50th Percentile of marks array:", percentile_50)

#75th Percentile of marks array
percentile_75 = np.percentile(marks, 75)
print("75th Percentile of marks array", percentile_75)
```

```python
#mean of marks array
mean_marks = np.mean(marks)

#Count the number of students who scored above the mean
above_mean_count = np.sum(marks > mean_marks)
print("Number of students who scored above the mean:", above_mean_count)
```
