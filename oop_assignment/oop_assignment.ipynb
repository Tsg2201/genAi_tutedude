```python
# Basic Class and Object Creation
class Product: #class definition
    def __init__(self,name,price,category): #constructor definition
        self.name=name
        self.price=price
        self.category=category
    
    def get_info(self): # getter function definition
        return f"product name is {self.name}, price is {self.price}, category is {self.category}"
    
    def apply_discount(self,percent): #optional function definition
        discounted_price=self.price-(self.price*percent/100)
        return f"Discounted Price of {self.name} is {discounted_price}"
```

```python
prod1=Product("laptop",1000,"electronics")
prod2=Product("Phone",500,"electronics")


prod1.get_info()
prod2.get_info()

prod1.apply_discount(10)
prod2.apply_discount(20)
```

```python
# Constructor and Encapsulation
 
class Product: #class definition
    def __init__(self,name,price,category): #constructor definition
        self.name=name
        self.__price=price
        self.category=category
    
    def get_info(self): # getter function definition
        return f"product name is {self.name}, price is {self.__price}, category is {self.category}"

    def get_price(self): #getter method to get price
        return (f"The price of {self.name} is {self.__price}")
    
    def set_price(self,price): #setter method to set price
        self.__price=price
```

```python
prod3=Product("Laptop",500,"Electronics")
print(prod3.get_info())
print(prod3.get_price())
prod3.set_price(1000)
print(prod3.get_price())
```

```python
class ElectronicPRoduct(Product):
    def __init__(self,name,price,category,warranty):
        super().__init__(name,price,category)
        self.warranty=warranty
    
    #Method Overiding
    def get_info(self):
        return(f"Product name is {self.name}, category is {self.category}, warranty is {self.warranty} years")
```

```python
smartwatch = ElectronicProduct("smart watch", 80000, "Electronics", 2)

# Demonstration
print(smartwatch.get_info())      # overridden method
print(smartwatch.get_price())     # inherited method
```

```python
#Polymorphism
class Laptop(Product):
    def __init__(self,name,price,category,warranty):
        super().__init__(name,price,category)
        self.warranty=warranty
    
    def get_info(self):
        return (f"Laptop name is {self.name}, category is {self.category}, warranty is {self.warranty} years")


class Mobile(Product):
    def __init__(self,name,price,category,storage):
        super().__init__(name,price,category)
        self.storage=storage
    
    def get_info(self):
        return (f"Mobile name is {self.name}, category is {self.category}, storage is {self.storage} GB")
```

```python
# Demonstration

l1=Laptop("Dell",50000,"Electronics",2)
m1=Mobile("Samsung",20000,"Electronics",128)

l2=Laptop("HP",60000,"Electronics",2)
m2=Mobile("Apple",80000,"Electronics",256)

products=[l1,m1,l2,m2]
laptops=[l1,l2]
mobiles=[m1,m2]
for l in laptops:
    print(l.get_info())

for m in mobiles:
    print(m.get_info())

for p in products:
    print(p.get_info())
```

```python
# Abstraction
from abc import ABC,abstractmethod
class Payment(ABC):
    @abstractmethod
    def pay(self):
        pass

class CreditCardPayment(Payment):
    def pay(self,amount):
        print("Processing Credit Card Payment")

 
class UPIPayment(Payment):
    def pay(self,amount):
        print("Processing UPI Payment")

 

 
```

```python
cc = CreditCardPayment()
upi = UPIPayment()

cc.pay(5000)
upi.pay(1500)
```

```python
# Magic methods and operator overloading
 
class Product: #class definition
    def __init__(self,name,price): #constructor definition
        self.name=name
        self.__price=price
       
    
    def get_info(self): # getter function definition
        return f"product name is {self.name}, price is {self.__price}"

    def get_price(self): #getter method to get price
        return (f"The price of {self.name} is {self.__price}")
    
    def set_price(self,price): #setter method to set price
        self.__price=price

    def __add__(self,other):
        return self.__price + other.__price

    def __str__ (self):
        return f"The combined price of {self.name} and {other.name} is {self.__price + other.__price}"
```

```python
prod1=Product("Laptop",50000)
prod2=Product("Mobile",20000)
total_price=prod1+prod2
print(total_price)
```

```python
# Mini Project



class Inventory:

    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, name):
        for p in self.products:
            if p.name == name:
                self.products.remove(p)
                break

    def get_total_value(self):
        total = 0
        for p in self.products:
            total += p.get_price()
        return total

    def show_all_products(self):
        for p in self.products:
            print(p.get_info())

class Store:

    def __init__(self, store_name):
        self.store_name = store_name
        self.inventory = Inventory()

    def add_new_product(self, name, price, category):
        product = Product(name, price, category)
        self.inventory.add_product(product)

    def show_summary(self):
        print("Store:", self.store_name)
        print("Products in inventory:")
        self.inventory.show_all_products()

        print("Total Inventory Value:", self.inventory.get_total_value())
```

```python
store = Store("Tech Store")

store.add_new_product("Laptop", 50000, "Electronics")
store.add_new_product("Phone", 20000, "Electronics")
store.add_new_product("Tablet", 30000, "Electronics")
 
```

```python
# Display Summary

store.show_summary()
```
