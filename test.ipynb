# Python OOP - Encapsulation

**Encapsulation** = Bundling data (attributes) and methods into a single unit (class), while restricting direct access to some internals.

### Key Concepts:
1. **Public** attributes — accessible anywhere
2. **Protected** (`_`) attributes — convention, should not be accessed outside class
3. **Private** (`__`) attributes — name-mangled, cannot be accessed directly
4. **Getters / Setters** — controlled access to private data
5. **`@property` decorator** — Pythonic way to define getters & setters

---
## 1. Basic Encapsulation — Public / Protected / Private

```python
class Employee:
    def __init__(self, name, salary, department):
        self.name = name              # public
        self._department = department  # protected (convention)
        self.__salary = salary        # private (name-mangled)

    # Getter for private attribute
    def get_salary(self):
        return self.__salary

    # Setter with validation
    def set_salary(self, new_salary):
        if new_salary < 0:
            raise ValueError("Salary cannot be negative!")
        self.__salary = new_salary

    def display(self):
        print(f"Employee: {self.name} | Dept: {self._department} | Salary: {self.__salary:,}")
```

```python
emp = Employee("Rahul", 75000, "Engineering")
emp.display()

# Public attribute - works fine
print(f"\nPublic  ->  emp.name = {emp.name}")

# Protected attribute - works but discouraged outside class
print(f"Protected ->  emp._department = {emp._department}")

# Private attribute - raises AttributeError
try:
    print(emp.__salary)
except AttributeError as e:
    print(f"Private ->  emp.__salary raised: {e}")

# Access via getter
print(f"Via getter ->  emp.get_salary() = {emp.get_salary():,}")
```

```python
class Animal():   

    def make_sound(self):
        pass


class Dog(Animal):
    def make_sound(self):
        print("Bark")


class Cat(Animal):
    def make_sound(self):
        print("Meow")


dog = Dog()
dog.make_sound()

c= Cat()
c.make_sound()
```

```python
# Update salary via setter
emp.set_salary(90000)
print(f"After raise -> salary = {emp.get_salary():,}")

# Try setting negative salary
try:
    emp.set_salary(-5000)
except ValueError as e:
    print(f"Negative salary -> {e}")

# Name mangling - still accessible but strongly discouraged
print(f"Name mangling -> emp._Employee__salary = {emp._Employee__salary:,}")
```

## 2. Encapsulation with `@property` Decorator

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.__owner = owner
        self.__balance = balance

    @property
    def balance(self):
        """Read the balance (getter)."""
        return self.__balance

    @balance.setter
    def balance(self, amount):
        """Set the balance with validation."""
        if amount < 0:
            raise ValueError("Balance cannot be negative!")
        self.__balance = amount

    @property
    def owner(self):
        return self.__owner

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive!")
        self.__balance += amount
        print(f"Deposited {amount:,}. New balance: {self.__balance:,}")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive!")
        if amount > self.__balance:
            raise ValueError("Insufficient funds!")
        self.__balance -= amount
        print(f"Withdrew {amount:,}. New balance: {self.__balance:,}")

    def __str__(self):
        return f"BankAccount(owner={self.__owner}, balance={self.__balance:,})"
```

```python
acc = BankAccount("Priya", 10000)
print(acc)

acc.deposit(5000)
acc.withdraw(3000)
print(f"Balance via property: {acc.balance:,}")

# Set balance directly via property setter
acc.balance = 50000
print(f"After setting balance to 50000: {acc.balance:,}")

# Try setting negative balance
try:
    acc.balance = -100
except ValueError as e:
    print(f"Negative balance -> {e}")

# Try over-withdrawing
try:
    acc.withdraw(999999)
except ValueError as e:
    print(f"Over-withdraw -> {e}")
```

## 3. Encapsulation with Validation — Student Class

```python
class Student:
    def __init__(self, name, age, grade):
        self.__name = name
        self.age = age      # goes through @age.setter
        self.grade = grade   # goes through @grade.setter

    @property
    def name(self):
        return self.__name

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if not isinstance(value, int) or value < 5 or value > 25:
            raise ValueError("Age must be an integer between 5 and 25.")
        self.__age = value

    @property
    def grade(self):
        return self.__grade

    @grade.setter
    def grade(self, value):
        valid_grades = ['A', 'B', 'C', 'D', 'F']
        if value.upper() not in valid_grades:
            raise ValueError(f"Grade must be one of {valid_grades}")
        self.__grade = value.upper()

    def __str__(self):
        return f"Student(name={self.__name}, age={self.__age}, grade={self.__grade})"
```

```python
s = Student("Ananya", 18, "A")
print(s)

# Valid updates
s.age = 20
s.grade = "b"
print(f"After updates: {s}")

# Invalid age
try:
    s.age = 2
except ValueError as e:
    print(f"Invalid age -> {e}")

# Invalid grade
try:
    s.grade = "Z"
except ValueError as e:
    print(f"Invalid grade -> {e}")

# Name is read-only (no setter)
try:
    s.name = "New Name"
except AttributeError as e:
    print(f"Read-only name -> {e}")
```

## Summary

| Feature | Syntax | Accessible outside class? |
|---|---|---|
| **Public** | `self.name` | Yes |
| **Protected** | `self._name` | Yes, but discouraged |
| **Private** | `self.__name` | No (name-mangled to `_ClassName__name`) |
| **Getter/Setter** | Methods or `@property` | Controlled access |
