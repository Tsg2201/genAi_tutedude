```python
sales = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue': [1200, 1500, 900, 2000, 1800]
}

sales_df = pd.DataFrame(sales)
sales_df
```

```python
print("Total revenue:", sales_df['Revenue'].sum())
```

```python
print("Average revenue:", sales_df['Revenue'].mean())
```

```python
sales_df[sales_df['Revenue'] == sales_df['Revenue'].max()]
```

```python
sales_df[sales_df['Revenue'] > sales_df['Revenue'].mean()]
```

```python
sales_df.plot(x='Day', y='Revenue', kind='line')
```
