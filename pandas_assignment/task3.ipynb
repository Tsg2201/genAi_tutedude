```python
marks = [78, 85, 90, 66, 72]
series = pd.Series(marks)
series
```

```python
print("Maximum marks:", series.max())
print("Minimum marks:", series.min())
print("Sum of marks:", series.sum())
print("Mean marks:", series.mean())
```

```python
result = series.apply(lambda x: x >= 70)
print(result)
```

```python
passed_count = (series >= 70).sum()
print("Number of students passed:", passed_count)
```
