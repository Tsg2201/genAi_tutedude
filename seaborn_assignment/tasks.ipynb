```python
# Install if needed
!pip install seaborn pandas matplotlib

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

 

# Load dataset
df = pd.read_csv("students.csv")  # <-- change path if needed

df.head()
```

## Task 1: Relational Plot

```python
# Relational plot (scatter default)
sns.relplot(data=df, x="math score", y="reading score", hue="gender")
plt.show()

# Scatter style explicitly
sns.relplot(data=df, x="math score", y="reading score", hue="gender", kind="scatter")
plt.show()
```

## Task 2: Line Plot as Scatter & Facet

```python
# Line plot
sns.lineplot(data=df, x="math score", y="writing score")
plt.show()

# Scatter-style equivalent
sns.scatterplot(data=df, x="math score", y="writing score")
plt.show()

# Faceting
sns.relplot(data=df, x="math score", y="writing score",
            col="gender", kind="scatter")
plt.show()
```

## Task 3: Distribution Plots

```python
# Histogram
sns.histplot(df["math score"], bins=20)
plt.show()

# KDE
sns.kdeplot(df["math score"], fill=True)
plt.show()

# Rug plot
sns.rugplot(df["math score"])
plt.show()

# Histogram + KDE
sns.histplot(df["math score"], kde=True)
plt.show()
```

## Task 4: Bivariate Distribution

```python
# Bivariate histogram
sns.histplot(data=df, x="math score", y="reading score")
plt.show()

# Bivariate KDE
sns.kdeplot(data=df, x="math score", y="reading score", fill=True)
plt.show()
```

## Task 5: Matrix Plots

```python
# Pairplot
sns.pairplot(df, hue="gender")
plt.show()

# Correlation heatmap
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

## Task 6: Categorical Plots

```python
# Bar plot
sns.barplot(data=df, x="gender", y="math score")
plt.show()

# Box plot
sns.boxplot(data=df, x="gender", y="math score")
plt.show()

# Violin plot
sns.violinplot(data=df, x="gender", y="math score")
plt.show()

# Count plot
sns.countplot(data=df, x="gender")
plt.show()
```

## Task 7: Regression Plots

```python
# Regression plot
sns.regplot(data=df, x="math score", y="reading score")
plt.show()

# lmplot with hue
sns.lmplot(data=df, x="math score", y="reading score", hue="gender")
plt.show()
```

## Task 8: Multi-Plots & Figure-Level

```python
# FacetGrid
g = sns.FacetGrid(df, col="gender")
g.map(sns.scatterplot, "math score", "reading score")
plt.show()

# relplot
sns.relplot(data=df, x="math score", y="reading score", hue="gender")
plt.show()

# catplot
sns.catplot(data=df, x="gender", y="math score", kind="box")
plt.show()

# displot
sns.displot(df["math score"], kde=True)
plt.show()
```
